# Introduction
Earthquakes are not strong for the violent movement of the soil caused by the sudden movement of rock material below the surface. Earthquakes start at tectonic plate boundaries. The hypocentre is the point on Earth where the earthquake begins, sometimes called the hypocentre, and the point on the Earth's surface just above the center of focus is called the epicenter. We can measure the strength of an earthquake in two ways: magnitude and magnitude. The magnitude is directly proportional to the energy released at the focal point of the earthquake. It is calculated from earthquakes recorded by an instrument called a seismograph. It is represented by Arabic numerals (such as 4.8, 9.0). Intensity is how strongly an earthquake is seen and felt by people in an area. It is a numerical measurement based on the relationship between people, objects, environments, and structures in the environment. Intensity is usually higher near the epicenter. It is represented by Roman numerals (eg.G. II, IV, XI). In the Philippines, the magnitude of earthquakes is determined using the PHIVOLCS Earthquake Intensity Scale (PEIS). There are two types of earthquakes: tectonic and volcanic. Tectonic earthquakes are caused by sudden movements along faults and plate boundaries.
Earthquakes caused by lava or magma rising from under a mighty volcano are called volcanic earthquakes. Currently, PHIVOLCS operates 108 (until December 2022) seismic monitoring stations across the Philippines. These stations are equipped with seismographs that detect and record earthquakes. Data is sent to the PHIVOLCS Data Retrieval Center (DRC) to determine seismic parameters such as magnitude, depth of focus, and epicenter. When earthquake data are determined, they will be published with the regional disaster announcement.

```python
#a.Looking at the Data Structure:
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv(path_db)

# Display the data structure
data.head()
```

```python
#b. Getting information/description of data:
# Number of rows and columns
num_rows, num_cols = data.shape
print("Number of rows:", num_rows)
print("Number of columns:", num_cols)

# Number and names of features' type
feature_types = data.dtypes.value_counts()
print("\nNumber and names of features' type:")
print(feature_types)

# Number of non-null values
non_null_counts = data.notnull().sum()
print("\nNumber of non-null values:")
print(non_null_counts)

# Name of features
features = data.columns.tolist()
print("\nName of features:")
print(features)
```

```python
#c.Understand the shape of the data (Histograms, box plots, etc.):
numeric_columns = ['Depth', 'Magnitude'] 

for col in numeric_columns:
    # Histogram
    data[col].plot(kind='hist', edgecolor='black')
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.title(f"Histogram of {col}")
    plt.show()
```

```python
# Box plot
for col in numeric_columns:
    data[col].plot(kind='box', vert=False) # vert=False for horizontal box plot
    plt.xlabel("Values")
    plt.title(f"Box Plot of {col}")
    plt.show()
```

```python

# 4. Data Cleaning - Identify missing values 

missing_values = data.isnull().sum()

print("Missing values:\n", missing_values)

# depth, latitude, longtiude , magnitude looks fine, which we will use
```

```python
# 4. Data Cleaning - Identify outliers
from scipy import stats
import numpy as np

# Specify the columns to analyze
columns = ["Depth", "Latitude", "Longitude", "Magnitude"]

for column in columns:
    mean = np.mean(data[column])
    print("Mean", mean)   
    std_dev = np.std(data[column])
    print("Standard Deviation:", std_dev)



# Identify outliers using IQR method for each column
outliers = {}

for column in columns:
    q1 = np.percentile(data[column], 25)
    q3 = np.percentile(data[column], 75)
    iqr = q3 - q1
    
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    column_outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)][column]
    outliers[column] = column_outliers

# Print the outliers for each column
for column, outlier_values in outliers.items():
    print("Possible Outliers", column, ":")
    print(outlier_values)
    print()


threshold = 10  # Z-score threshold for outlier detection
z_scores = [(x - mean) / std_dev for x in data[column]]
outliers2 = [data[column][i] for i, z in enumerate(z_scores) if abs(z) > threshold]

print("Possible Outliers2", outliers2)
```
